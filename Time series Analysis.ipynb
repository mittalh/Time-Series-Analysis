```python
import pandas as pd
from datetime import datetime
import numpy as np
date_rng = pd.date_range(start='1/1/2018',freq='10D',end='1/1/2019')
import random
random.seed(121)
```

```python
Region=['NorthEastern','SouthCentral']
State=['New York','Texas']
```

```python

df = pd.DataFrame(date_rng, columns=['date'])
random.seed(42)
df['Sales_'] = np.random.randint(100,1000,size=(len(date_rng)))    # Random Sales between 100 and 1000
random.seed(12)
df['Profit']=np.random.randint(-20,20,size=(len(date_rng)))     # Random Profit in Between -20 and 20(-ve --> Loss)
random.seed(12)
df['Region']=[random.randint(0,1) for j in range(len(date_rng))]
random.seed(12)
df['State']=[random.randint(0,1) for j in range(len(date_rng))]
```

```python
df[['Region']]=df[['Region']].replace({0:'NorthEastern',1:'SouthCentral'})
df[['State']]=df[['State']].replace({0:'New York',1:'Texas'})
```

```python
df.head()
```

```python
from matplotlib.pyplot import plot_date
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plot_date(df[df['Region']=='SouthCentral'].date,df[df['Region']=='SouthCentral'].Profit,linestyle='solid')
plot_date(df[df['Region']=='NorthEastern'].date,df[df['Region']=='NorthEastern'].Profit,linestyle='solid')
plt.legend(['SouthCentral','NorthEastern'])
plt.xlabel("Date")
plt.ylabel("Profit")
plt.tight_layout()




# Conclusion:
#SouthCentral: Profits are decreasing till 8th Month and then increasing 
# NorthEastern: Profits are increasing till 11th Month
```

```python
plt.style.use('seaborn')
plot_date(df[df['Region']=='SouthCentral'].date,df[df['Region']=='SouthCentral'].Sales_,linestyle='solid')
plot_date(df[df['Region']=='NorthEastern'].date,df[df['Region']=='NorthEastern'].Sales_,linestyle='solid')
plt.legend(['SouthCentral','NorthEastern'])
plt.xlabel("Date")
plt.ylabel("Sales_")
plt.tight_layout()



# Conclusion
#SouthCentral: Increasing till mid then Decreasing
# NorthEastern: No Trend
```
